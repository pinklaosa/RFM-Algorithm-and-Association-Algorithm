## set DOCNUM

```python
df = pd.read_csv(r'Data/DATASET_RFM.csv')

df.index = df.index.rename('')
df['DOCNUM_ID'] = df.index
re_df=df.reindex(columns= ['DOCNUM_ID','PEOPLE', 'STKCOD', 'DOCNUM','TRNQTY','STKDES'])
re_df
```

```python
df_re = df.reindex(columns=['DOCNUM'])
df_re
```

คัด DOCNUM ออกมาให้เป็นลำดับแบบไม่ซ้ำ

```python
df_re2 = df_re.drop_duplicates(subset='DOCNUM')
df_re2 = df_re2.reset_index()
df_re2.index += 1
df_re2.index = df_re2.index.rename('')
df_re2['DOCNUM_ID'] = df_re2.index
df_re3 = df_re2.reindex(columns=['DOCNUM_ID', 'DOCNUM'])
df_re3
```

ทำ loop เพื่อทำการ set DOCNUM ให้ให้สำหรับขั้นตอน AR

```python
x = 1
for col_idx,data in df_re3.iterrows():

    df88 = data["DOCNUM"]
    condition_1 = re_df['DOCNUM'] == df88
    re_df[ condition_1 ]
    df_set = re_df[ condition_1 ]
    df_set['DOCNUM_ID'] =x
    
    #x = x+1
    if(x==1):
        df_mi = df_set
        x = x+1
    else:
        df_mi = pd.concat([df_mi,df_set],ignore_index = True)
        df_mi.reset_index() 
        x=x+1
```

```python
df_mi
```

```python
df_mi.to_csv(r'Data/NORMAL_DATA_RFM.csv',encoding='utf-8',index=False)
```
