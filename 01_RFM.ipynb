## Dropping records with empty values

```python
df.dropna(axis=0,inplace=True)
df
```

## Change type TRNQTY to int 

```python
df['TRNQTY'] = df['TRNQTY'].astype(int)
df= df[['PEOPLE','DOCDAT','TRNQTY','NETVAL']].drop_duplicates()
df = df.reset_index().drop(columns='index')
df
```

# Filter required Columns

```python
df['DOCDAT'].min(),df['DOCDAT'].max()
```

```python
PRESENT = dt.datetime(2019,2,2)
df['DOCDAT'] = pd.to_datetime(df['DOCDAT'])
df.head()
```

## Start RFM

```python
rfm= df.groupby('PEOPLE').agg({'DOCDAT': lambda date: (PRESENT - date.max()).days,
                                    'TRNQTY': lambda num: sum(num),
                                    'NETVAL': lambda price: price.sum()})
```

```python
rfm.columns=['recency','frequency','monetary']
rfm.columns
```

```python
rfm.dropna(axis=0,inplace=True)
rfm = rfm[(rfm['frequency']>1)]
rfm
```

```python
rfm_score = rfm
rfm_score['r_quartile'] = pd.qcut(rfm['recency'], 4, ['1','2','3','4'])
rfm_score['f_quartile'] = pd.qcut(rfm['frequency'], 4, ['4','3','2','1'])
rfm_score['m_quartile'] = pd.qcut(rfm['monetary'], 4, ['4','3','2','1'])
rfm_score
```

# RFM Result Interpretation
-Combine all three quartiles(r_quartile,f_quartile,m_quartile) in a single column, this rank will help you to segment the customers well group.

```python
rfm_score['RFM_Score'] = rfm_score.r_quartile.astype(str)+ rfm_score.f_quartile.astype(str) + rfm_score.m_quartile.astype(str)
# rfm_score.to_csv(r'RFM_SCORE.csv')
rfm_score
```

```python
rfm_111 = rfm_score[rfm_score['RFM_Score']=='111'].sort_values('monetary', ascending=False)
```
